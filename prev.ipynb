Analisando Autorrelação SELIC e IPCA

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from bcb import sgs
sns.set()
```

```python
tabela = sgs.get(
    codes = {'selic': 4390,
             "ipca": 433},
    start = "2005-01-01",
    end = "2024-10-10"
).dropna()
tabela
```

```python
# ACF IPCA (Usa-se biblioteca statsmodels)
sm.tsa.stattools.acf(x = tabela.ipca.values, nlags=24)
```

Gráfico autocorrelação IPCA

```python
sm.graphics.tsa.plot_acf(x = tabela.ipca.values, lags=24);
```

Gráfico autocorrelação SELIC 

```python
sm.graphics.tsa.plot_acf(x = tabela.selic.values, lags=24);
```
